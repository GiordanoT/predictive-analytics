```python
import pandas as pd
from pathlib import Path  
```

```python
df = pd.read_csv('activities.csv', sep=';', index_col=0)
df
```

```python
# Building a new dataframe
new_df = pd.DataFrame()
new_df['Activity'] = df['Activity']
new_df['Subject'] = df['Subject']
```

```python
# Aggregate together data from sensor ACC captured in 5 different body's part.
new_df['ACC x'] = df['T(A, x)'] + df['RA(A, x)'] + df['LA(A, x)'] + df['RL(A, x)'] + df['LL(A, x)']
new_df['ACC y'] = df['T(A, y)'] + df['RA(A, y)'] + df['LA(A, y)'] + df['RL(A, y)'] + df['LL(A, y)']
new_df['ACC z'] = df['T(A, z)'] + df['RA(A, z)'] + df['LA(A, z)'] + df['RL(A, z)'] + df['LL(A, z)']
```

```python
# Aggregate together data from sensor GYR captured in 5 different body's part.
new_df['GYR x'] = df['T(G, x)'] + df['RA(G, x)'] + df['LA(G, x)'] + df['RL(G, x)'] + df['LL(G, x)']
new_df['GYR y'] = df['T(G, y)'] + df['RA(G, y)'] + df['LA(G, y)'] + df['RL(G, y)'] + df['LL(G, y)']
new_df['GYR z'] = df['T(G, z)'] + df['RA(G, z)'] + df['LA(G, z)'] + df['RL(G, z)'] + df['LL(G, z)']
```

```python
# Aggregate together data from sensor MAG captured in 5 different body's part.
new_df['MAG x'] = df['T(M, x)'] + df['RA(M, x)'] + df['LA(M, x)'] + df['RL(M, x)'] + df['LL(M, x)']
new_df['MAG y'] = df['T(M, y)'] + df['RA(M, y)'] + df['LA(M, y)'] + df['RL(M, y)'] + df['LL(M, y)']
new_df['MAG z'] = df['T(M, z)'] + df['RA(M, z)'] + df['LA(M, z)'] + df['RL(M, z)'] + df['LL(M, z)']
```

```python
# Deleting outliers capture by ACC
indexes = new_df.loc[new_df['ACC x'] > 50].index
new_df.drop(indexes, inplace=True)
indexes = new_df.loc[new_df['ACC x'] < -50].index
new_df.drop(indexes, inplace=True)

indexes = new_df.loc[new_df['ACC y'] > 50].index
new_df.drop(indexes, inplace=True)
indexes = new_df.loc[new_df['ACC y'] < -50].index
new_df.drop(indexes, inplace=True)

indexes = new_df.loc[new_df['ACC z'] > 50].index
new_df.drop(indexes, inplace=True)

indexes = new_df.loc[new_df['ACC z'] < -50].index
new_df.drop(indexes, inplace=True)
```

```python
# Deleting outliers capture by GYR
indexes = new_df.loc[new_df['GYR x'] > 10].index
new_df.drop(indexes, inplace=True)
indexes = new_df.loc[new_df['GYR x'] < -10].index
new_df.drop(indexes, inplace=True)

indexes = new_df.loc[new_df['GYR y'] > 10].index
new_df.drop(indexes, inplace=True)
indexes = new_df.loc[new_df['GYR y'] < -10].index
new_df.drop(indexes, inplace=True)

indexes = new_df.loc[new_df['GYR z'] > 10].index
new_df.drop(indexes, inplace=True)
indexes = new_df.loc[new_df['GYR z'] < -10].index
new_df.drop(indexes, inplace=True)
```

```python
# Checking if there are NaN values.
new_df.isnull().any()
```

```python
# Saving the cleaned dataframe, so we can easly use it.
filepath = Path('activities_cleaned.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
new_df.to_csv(filepath, sep=';')
```
