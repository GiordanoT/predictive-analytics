```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('activities_cleaned.csv', sep=';', index_col=0)
df
```

```python
# Drawing the distribution.
%matplotlib inline
df.hist(column=['ACC x', 'ACC y', 'ACC z'], figsize=(7, 5))
df.hist(column=['GYR x', 'GYR y', 'GYR z'], figsize=(7, 5))
df.hist(column=['MAG x', 'MAG y', 'MAG z'], figsize=(7, 5))
```

```python
# Drawing the boxplot.
%matplotlib inline
df.boxplot(column=['ACC x', 'ACC y', 'ACC z'], figsize=(10, 5))
```

```python
# Drawing the boxplot.
%matplotlib inline
df.boxplot(column=['GYR x', 'GYR y', 'GYR z'], figsize=(10, 5))
```

```python
# Drawing the boxplot.
%matplotlib inline
df.boxplot(column=['MAG x', 'MAG y', 'MAG z'], figsize=(10, 5))
```

```python
# Drawing the Correlation Matrix.
corr_matrix = df.drop(columns=['Activity', 'Subject']).corr()
sn.heatmap(corr_matrix, annot=False)
plt.show()
```
