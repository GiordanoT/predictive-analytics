```python
import pandas as pd
from pathlib import Path  
```

```python
df = pd.read_csv('activities.csv', sep=';', index_col=0)
```

```python
new_df = pd.DataFrame()
```

```python
new_df['Activity'] = df['Activity']
new_df['Subject'] = df['Subject']
```

```python
new_df['ACC x'] = df['T(A, x)'] + df['RA(A, x)'] + df['LA(A, x)'] + df['RL(A, x)'] + df['LL(A, x)']
new_df['ACC y'] = df['T(A, y)'] + df['RA(A, y)'] + df['LA(A, y)'] + df['RL(A, y)'] + df['LL(A, y)']
new_df['ACC z'] = df['T(A, z)'] + df['RA(A, z)'] + df['LA(A, z)'] + df['RL(A, z)'] + df['LL(A, z)']
```

```python
new_df['GYR x'] = df['T(G, x)'] + df['RA(G, x)'] + df['LA(G, x)'] + df['RL(G, x)'] + df['LL(G, x)']
new_df['GYR y'] = df['T(G, y)'] + df['RA(G, y)'] + df['LA(G, y)'] + df['RL(G, y)'] + df['LL(G, y)']
new_df['GYR z'] = df['T(G, z)'] + df['RA(G, z)'] + df['LA(G, z)'] + df['RL(G, z)'] + df['LL(G, z)']
```

```python
new_df['MAG x'] = df['T(M, x)'] + df['RA(M, x)'] + df['LA(M, x)'] + df['RL(M, x)'] + df['LL(M, x)']
new_df['MAG y'] = df['T(M, y)'] + df['RA(M, y)'] + df['LA(M, y)'] + df['RL(M, y)'] + df['LL(M, y)']
new_df['MAG z'] = df['T(M, z)'] + df['RA(M, z)'] + df['LA(M, z)'] + df['RL(M, z)'] + df['LL(M, z)']
```

```python
new_df['ACC x'] = round(new_df['ACC x'] / 5, 4)
new_df['ACC y'] = round(new_df['ACC y'] / 5, 4)
new_df['ACC z'] = round(new_df['ACC z'] / 5, 4)

new_df['GYR x'] = round(new_df['GYR x'] / 5, 4)
new_df['GYR y'] = round(new_df['GYR y'] / 5, 4)
new_df['GYR z'] = round(new_df['GYR z'] / 5, 4)

new_df['MAG x'] = round(new_df['MAG x'] / 5, 4)
new_df['MAG y'] = round(new_df['MAG y'] / 5, 4)
new_df['MAG z'] = round(new_df['MAG z'] / 5, 4)
```

```python
new_df
```

```python
filepath = Path('activities_cleaned.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
new_df.to_csv(filepath, sep=';')
```
