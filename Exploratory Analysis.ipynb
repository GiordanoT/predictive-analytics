```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('activities_cleaned.csv', sep=';', index_col=0)
df
```

```python
%matplotlib inline
df.hist(column=['ACC x', 'ACC y', 'ACC z'], figsize=(6.5, 6.5))
df.hist(column=['GYR x', 'GYR y', 'GYR z'], figsize=(6.5, 6.5))
df.hist(column=['MAG x', 'MAG y', 'MAG z'], figsize=(6.5, 6.5))
```

```python
indexes = df.loc[df['ACC x'] > 50].index
df.drop(indexes, inplace=True)
indexes = df.loc[df['ACC x'] < -50].index
df.drop(indexes, inplace=True)

indexes = df.loc[df['ACC y'] > 50].index
df.drop(indexes, inplace=True)
indexes = df.loc[df['ACC y'] < -50].index
df.drop(indexes, inplace=True)

indexes = df.loc[df['ACC z'] > 50].index
df.drop(indexes, inplace=True)

indexes = df.loc[df['ACC z'] < -50].index
df.drop(indexes, inplace=True)
```

```python
indexes = df.loc[df['GYR x'] > 10].index
df.drop(indexes, inplace=True)
indexes = df.loc[df['GYR x'] < -10].index
df.drop(indexes, inplace=True)

indexes = df.loc[df['GYR y'] > 10].index
df.drop(indexes, inplace=True)
indexes = df.loc[df['GYR y'] < -10].index
df.drop(indexes, inplace=True)

indexes = df.loc[df['GYR z'] > 10].index
df.drop(indexes, inplace=True)
indexes = df.loc[df['GYR z'] < -10].index
df.drop(indexes, inplace=True)
```

```python
%matplotlib inline
df.hist(column=['ACC x', 'ACC y', 'ACC z'], figsize=(6.5, 6.5))
df.hist(column=['GYR x', 'GYR y', 'GYR z'], figsize=(6.5, 6.5))
df.hist(column=['MAG x', 'MAG y', 'MAG z'], figsize=(6.5, 6.5))
```

```python
%matplotlib inline
df.boxplot(column=['ACC x', 'ACC y', 'ACC z'], figsize=(10, 5))
df.boxplot(column=['GYR x', 'GYR y', 'GYR z'], figsize=(10, 5))
df.boxplot(column=['MAG x', 'MAG y', 'MAG z'], figsize=(10, 5))
```

```python
corr_matrix = df.drop(columns=['Activity', 'Subject']).corr()
sn.heatmap(corr_matrix, annot=False)
plt.show()
```
