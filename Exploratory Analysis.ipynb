```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('activities_cleaned.csv', sep=';', index_col=0)
```

```python
%matplotlib inline
df.groupby(by=['Activity', 'Subject']).mean().hist(column=['ACC x', 'ACC y', 'ACC z'], figsize=(6.5, 6.5))
```

```python
%matplotlib inline
df.groupby(by=['Activity', 'Subject']).mean().hist(column=['GYR x', 'GYR y', 'GYR z'], figsize=(6.5, 6.5))
```

```python
%matplotlib inline
df.groupby(by=['Activity', 'Subject']).mean().hist(column=['MAG x', 'MAG y', 'MAG z'], figsize=(6.5, 6.5))
```

```python
%matplotlib inline
df.groupby(by=['Activity', 'Subject']).mean().boxplot(figsize=(10, 5))
```

```python
corr_matrix = df.groupby(by=['Activity', 'Subject']).mean().corr()
sn.heatmap(corr_matrix, annot=False)
plt.show()
```
